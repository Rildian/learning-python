```python
import pandas as pd
import re
```

```python
dados = pd.read_csv('house.csv')
dados
```

```python
filter1 = dados.loc[(dados['Rooms'] == 3) & (dados['Type'] == 'h') & (dados['Price'] <= 500000)] # where em ação
filter1
```

```python
filter2 = dados.loc[dados['Address'].str.contains('Turner St')] # o "in" padrão do python
filter2
```

```python
filtro3 = dados.loc[dados['Address'].str.contains('turner st', flags=re.I)]
filtro3
```

```python
filter4 = dados.loc[(dados['Address'].str.contains('^59', flags=re.I)) & (dados['Price'] < 500000)] # esse ^ garante que começa com 59
filter4
```

```python
dados.loc[dados['SellerG'] == 'Nelson', 'SellerG'] = 'Ronaldo' # this method changes on this column for this value
dados
```

```python
dados.loc[dados['SellerG'] == 'Ronaldo', ['Method', 'Regionname']] = 'Teste' # onde o seller for ronaldo, na coluna "method" troque para P, n é mto intuitivo
dados
```

```python
apresentar = dados.groupby(['SellerG']).sum(['Price'])
apresentar.sort_values(['Price'], ascending=False).head(10) # isso aq ve o top 10 vendedores baseado no PREÇO
```

```python
aprsentar = dados.groupby(['SellerG']).sum(['Rooms']) # note que sum é um método do group by, então precisa ficar groupby().sum(), sacou?
apresentar.sort_values(['Rooms'], ascending=False).head(10)
```
